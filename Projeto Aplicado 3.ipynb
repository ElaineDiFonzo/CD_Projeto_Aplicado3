## Ciência de dados
</div>

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Carregamento das bases
funcionarios_path = 'funcionarios_sintetico.csv'
vagas_path = 'vagas_sintetico.csv'

# Criação dos datasets
funcionarios_df = pd.read_csv(funcionarios_path, encoding='latin1', delimiter=';')
vagas_df = pd.read_csv(vagas_path, encoding='latin1', delimiter=';')

# Limpeza dos dados: Foi Removido colunas desnecessárias
funcionarios_df = funcionarios_df.drop(columns=['Unnamed: 9'], errors='ignore')
vagas_df = vagas_df.drop(columns=['Unnamed: 7'], errors='ignore')

# Limpeza e normalização das colunas habilidades e requisitos
funcionarios_df['Habilidades'] = funcionarios_df['Habilidades'].str.replace('Ã', 'ã').str.strip()
vagas_df['Requisitos'] = vagas_df['Requisitos'].str.strip()

# Preparação dos dados para aplicação de TF-IDF
funcionarios_habilidades = funcionarios_df['Habilidades'].fillna('')
vagas_requisitos = vagas_df['Requisitos'].fillna('')

# Inicialização do vetorizador TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Ajuste e transformação dos dados de habilidades e requisitos
tfidf_funcionarios = tfidf_vectorizer.fit_transform(funcionarios_habilidades)
tfidf_vagas = tfidf_vectorizer.transform(vagas_requisitos)

# Calculo de similaridade de cosseno entre funcionários e vagas
cosine_similarities = cosine_similarity(tfidf_funcionarios, tfidf_vagas)

# Gerar recomendações e Top 5 funcionários para a vaga
recommendations = []
for vaga_idx, vaga_row in vagas_df.iterrows():
    similarity_scores = cosine_similarities[:, vaga_idx]
    top_employees = similarity_scores.argsort()[::-1][:5]
    for emp_idx in top_employees:
        recommendations.append({
            'ID_Vaga': vaga_row['ID'],
            'Nome_Vaga': vaga_row['Nome da Vaga'],
            'Matrícula_Funcionário': funcionarios_df.loc[emp_idx, 'Matrícula'],
            'Nome_Funcionário': funcionarios_df.loc[emp_idx, 'Nome'],
            'Similaridade': similarity_scores[emp_idx]
        })

# Conversão das recomendações em um DataFrame
recommendations_df = pd.DataFrame(recommendations)
print(recommendations_df)
```

```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt

# Carregar os dados
funcionarios_path = 'funcionarios_sintetico.csv'  # Substitua pelo caminho correto
vagas_path = 'vagas_sintetico.csv'

# Carregar os datasets
funcionarios_df = pd.read_csv(funcionarios_path, encoding='latin1', delimiter=';')
vagas_df = pd.read_csv(vagas_path, encoding='latin1', delimiter=';')

# Limpeza de dados: Remover colunas desnecessárias
funcionarios_df = funcionarios_df.drop(columns=['Unnamed: 9'], errors='ignore')
vagas_df = vagas_df.drop(columns=['Unnamed: 7'], errors='ignore')

# Normalizar e limpar as colunas de habilidades e requisitos
funcionarios_df['Habilidades'] = funcionarios_df['Habilidades'].str.replace('Ã', 'ã').str.strip()
vagas_df['Requisitos'] = vagas_df['Requisitos'].str.strip()

# Preparar os dados para TF-IDF
funcionarios_habilidades = funcionarios_df['Habilidades'].fillna('')
vagas_requisitos = vagas_df['Requisitos'].fillna('')

# Inicializar o vetorizador TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Ajustar e transformar os dados de habilidades e requisitos
tfidf_funcionarios = tfidf_vectorizer.fit_transform(funcionarios_habilidades)
tfidf_vagas = tfidf_vectorizer.transform(vagas_requisitos)

# Calcular similaridade de cosseno entre funcionários e vagas
cosine_similarities = cosine_similarity(tfidf_funcionarios, tfidf_vagas)

# Gerar recomendações
recommendations = []
for vaga_idx, vaga_row in vagas_df.iterrows():
    similarity_scores = cosine_similarities[:, vaga_idx]
    top_employees = similarity_scores.argsort()[::-1][:5]  # Top 5 funcionários para a vaga
    for emp_idx in top_employees:
        recommendations.append({
            'ID_Vaga': vaga_row['ID'],
            'Nome_Vaga': vaga_row['Nome da Vaga'],
            'Matrícula_Funcionário': funcionarios_df.loc[emp_idx, 'Matrícula'],
            'Nome_Funcionário': funcionarios_df.loc[emp_idx, 'Nome'],
            'Similaridade': similarity_scores[emp_idx]
        })

# Converter as recomendações em um DataFrame
recommendations_df = pd.DataFrame(recommendations)

# Relevância para métricas
relevance_threshold = 0.7
recommendations_df['Relevant'] = recommendations_df['Similaridade'] >= relevance_threshold

# Cálculos Precision, Recall e F1-Score
true_positives = recommendations_df['Relevant'].sum()
false_positives = len(recommendations_df) - true_positives
false_negatives = vagas_df.shape[0] - true_positives  # Assumindo uma correspondência relevante por vaga

precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)
f1_score = 2 * (precision * recall) / (precision + recall)

metrics = {
    "Precision": precision,
    "Recall": recall,
    "F1-Score": f1_score,
    "True Positives": true_positives,
    "False Positives": false_positives,
    "False Negatives": false_negatives,
}

# Exibir métricas
print("Métricas de Avaliação:")
for key, value in metrics.items():
    print(f"{key}: {value:.2f}")

# Visualização: Curva de Precisão
thresholds = np.linspace(0, 1, 50)
precisions = [recommendations_df['Similaridade'].ge(t).mean() for t in thresholds]
recalls = [recommendations_df['Similaridade'].ge(t).sum() / true_positives for t in thresholds]

plt.figure(figsize=(8, 6))
plt.plot(recalls, precisions, label="Curva de Precisão")
plt.xlabel("Revocação")
plt.ylabel("Precisão")
plt.title("Curva de Precisão")
plt.legend()
plt.grid()
plt.show()
```

```
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc

#  Curvas de precisão revocação
precision, recall, _ = precision_recall_curve(recommendations_df['Relevant'], recommendations_df['Similaridade'])
plt.figure(figsize=(6, 4))
plt.plot(recall, precision, label="Curvas de precisão revocação")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Curvas de precisão revocação")
plt.legend()
plt.grid()
plt.show()

# Curvas ROC
fpr, tpr, _ = roc_curve(recommendations_df['Relevant'], recommendations_df['Similaridade'])
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"Curvas ROC (AUC = {roc_auc:.2f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curvas ROC")
plt.legend()
plt.grid()
plt.show()
```
